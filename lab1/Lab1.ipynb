```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes_X_train = diabetes.data[:-20]
diabetes_X_test = diabetes.data[-20:]
```

```python
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]
```

```python
regr = LinearRegression()
```

```python
regr.fit(diabetes_X_train, diabetes_y_train)
```

```python
#Exercise 2.1
#Yes, using all the available features improves the estimator's performance
print("Residual sum of squares: %.2f" 
      % np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2))
```

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(estimator=regr, n_features_to_select=2, step=1)
rfe.fit(diabetes_X_train,diabetes_y_train)
print rfe.ranking_
```

```python
#Exercise 2.2
print("Residual sum of squares: %.2f"
      % np.mean((rfe.predict(diabetes_X_test) - diabetes_y_test) ** 2))
```

```python
# 3 is the optimal number of features to be used
X = []
y = []
# Sum 0 to 99
s = 0
for x in range(1, 11):
    rfe = RFE(estimator=regr, n_features_to_select=x, step=1)
    rfe.fit(diabetes_X_train,diabetes_y_train)
    print("Residual sum of squares: %.2f"
      % np.mean((rfe.predict(diabetes_X_test) - diabetes_y_test) ** 2))
    X.append(x)
    y.append(np.mean((rfe.predict(diabetes_X_test) - diabetes_y_test) ** 2))
plt.xlabel('no of features')
plt.ylabel('mean squared error')
plt.title('# features vs. rss')
plt.plot(X,y)
plt.show()
```

```python
# Exercise 4 - Part 1
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
(rows, columns) = iris.data.shape

iris_X_train = iris.data[:int(floor(rows*(-0.2)))]
iris_X_test = iris.data[int(floor(rows*(-0.2))):]

iris_y_train = iris.target[:int(floor(rows*(-0.2)))]
iris_y_test = iris.target[int(floor(rows*(-0.2))):]
```

```python
# Create linear SVM

from sklearn import svm
clf_linear = svm.SVC(kernel='linear')
clf_linear.fit(iris_X_train, iris_y_train)

pred = clf_linear.predict(iris_X_test)

errors = 0.0
for i in range(len(pred)):
    if pred[i] != iris_y_test[i]:
        errors += 1.0
print "Test data errors = ",errors, "out of", len(pred), "predictions"

total_pred = clf_linear.predict(iris.data)
errors = 0.0
for i in range(len(total_pred)):
    if total_pred[i] != iris.target[i]:
        errors += 1.0
print "Total errors = ",errors, "out of", len(total_pred), "predictions"
```

```python
# Create non-linear SVM using the Radial Basis Kernel function

clf_non_linear = svm.SVC(kernel='rbf')
clf_non_linear.fit(iris_X_train,iris_y_train)

pred = clf_non_linear.predict(iris_X_test)
errors = 0.0
for i in range(len(pred)):
    if pred[i] != iris_y_test[i]:
        errors += 1.0
print "Test data errors = ",errors, "out of", len(pred), "predictions"

# Calculate prediction error for all points
total_pred_nl = clf_non_linear.predict(iris.data)
errors = 0.0
for i in range(len(total_pred_nl)):
    if total_pred_nl[i] != iris.target[i]:
        errors += 1.0
print "Total errors = ",errors, "out of", len(total_pred_nl), "predictions"
```

```python
# Part 2
from sklearn.tree import DecisionTreeClassifier as dtree
tree = dtree()
tree.fit(iris_X_train,iris_y_train)
         
pred = tree.predict(iris_X_test)
errors = 0.0
for i in range(len(pred)):
    if pred[i] != iris_y_test[i]:
        errors += 1.0
print "Test data errors = ",errors, "out of", len(pred), "predictions"

total_pred_tree = tree.predict(iris.data)
errors = 0.0
for i in range(len(total_pred_tree)):
    if total_pred_tree[i] != iris.target[i]:
        errors += 1.0
print "Total errors = ",errors, "out of", len(total_pred_tree), "predictions"
```

```python
# Exercise 5
from sklearn.cluster import KMeans
from sklearn import metrics

kmeans = KMeans(init='random', n_clusters = 3, n_init=10)
kmeans.fit(iris.data)

# Homogeneity score: evaluates if each cluster contains only memebers of a single class
print "Homogeneity_score (kmeans) = ",metrics.homogeneity_score(iris.target,kmeans.labels_)
print "Homogeneity_score (non-linear SVM) = ",metrics.homogeneity_score(iris.target,total_pred_nl)
print "Homogeneity_score (decision tree) = ",metrics.homogeneity_score(iris.target,total_pred_tree)

print 

# Completeness score: evaluates if all members of a given class are assigned to the same cluster
print "Completeness_score (kmeans) = ",metrics.completeness_score(iris.target,kmeans.labels_)
print "Completeness_score (non-linear SVM) = ",metrics.completeness_score(iris.target,total_pred_nl)
print "Completeness_score (decision tree) = ",metrics.completeness_score(iris.target,total_pred_tree)
```
